```python
df2 = df[['popularity','durationms','danceability', 'acousticness','energy', 'liveness', 'loudness', 'speechiness','tempo', 'valence','genre']]
```

```python
df.to_csv("initial_df_classification.csv")
df2.to_csv("final_df_classification.csv")
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df2, test_size=0.2)
```

```python
y_train = train['genre']
y_test = test['genre']
X_train = train.drop('genre',axis=1)
X_test = test.drop('genre',axis=1)
```

```python
XT = pd.DataFrame(X_train,columns=df2.iloc[:,:-1].columns)
XTe = pd.DataFrame(X_test,columns=df2.iloc[:,:-1].columns)
XT.to_csv('Trainingset.csv')
XTe.to_csv('Testingset.csv')
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
# scaler = StandardScaler()
# X_train = scaler.fit_transform(X_train)
# X_test = scaler.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB, GaussianNB
MyModelNB= GaussianNB(var_smoothing=0)
MyModelNB.fit(X_train, y_train)
Prediction = MyModelNB.predict(X_test)
cnf_matrix = confusion_matrix(y_test, Prediction)
print("\nThe confusion matrix is:")
disp = ConfusionMatrixDisplay(confusion_matrix=cnf_matrix,display_labels=MyModelNB.classes_)  
disp.plot()
plt.show()
```

```python
print(classification_report(y_test,Prediction))
```

```python
print(np.round(MyModelNB.predict_proba(X_test),2))
```

```python
from sklearn.naive_bayes import MultinomialNB, GaussianNB
MyModelNB= MultinomialNB()
MyModelNB.fit(X_train, y_train)
Prediction = MyModelNB.predict(X_test)
cnf_matrix = confusion_matrix(y_test, Prediction)
print("\nThe confusion matrix is:")
disp = ConfusionMatrixDisplay(confusion_matrix=cnf_matrix,display_labels=MyModelNB.classes_)  
disp.plot()
plt.show()
print(classification_report(y_test,Prediction))
```

#Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=9,criterion='entropy',random_state=0)
clf.fit(X_train,y_train)
```

```python
pred = clf.predict(X_test)
```

```python
cnf_matrix = confusion_matrix(y_test, pred)
print("\nThe confusion matrix is:")
disp = ConfusionMatrixDisplay(confusion_matrix=cnf_matrix,display_labels=clf.classes_)  
disp.plot()
plt.show()
```

```python
print(classification_report(y_test,pred))
```
