```python
import pandas as pd 
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/MLEXAM1/final_edm_dataset.csv')
#df2 = pd.read_csv('/content/drive/MyDrive/MLEXAM1/final_pop_dataset.csv')
df3 = pd.read_csv('/content/drive/MyDrive/MLEXAM1/final_rock_dataset.csv')
df4 = pd.read_csv('/content/drive/MyDrive/MLEXAM1/final_rap_dataset.csv')
df5 = pd.read_csv('/content/drive/MyDrive/MLEXAM1/final_country_dataset.csv')
```

```python
df1['genre'] = 'edm'
#df2['genre'] = 'pop'
df3['genre'] = 'rock'
df4['genre'] = 'rap'
df5['genre'] = 'country'
#df = pd.concat([df1,df2,df3,df4,df5])
df = pd.concat([df1,df3,df4,df5])
```

```python
df
```

```python
df2 = df[['popularity','durationms','danceability', 'acousticness','energy', 'liveness', 'loudness', 'speechiness','tempo', 'valence']]
```

```python
df2
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(df2.corr(), annot=True)
```

```python
df2 = df2[['loudness','energy']]
```

```python
plt.scatter(df2['energy'],df2['loudness'])
plt.xlabel('energy')
plt.ylabel('loudness')
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df2, test_size=0.2,random_state=21)
```

```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
model1 = smf.ols(formula='loudness ~ energy', data=train).fit()
```

```python
print(model1.summary())
```

```python

p = model1.params

# generate x-values for your regression line (two is sufficient)
x = np.arange(1, 10)/10

# scatter-plot data
ax = df2.plot(x='energy', y='loudness', kind='scatter')

# plot regression line on the same axes, set x-axis limits
ax.plot(x, p.Intercept + p.energy * x, color='red')
```

```python
from sklearn.metrics import mean_squared_error

Y_pred = model1.predict(test['energy'])

print("MSPE:",mean_squared_error(test['loudness'],Y_pred))
```
